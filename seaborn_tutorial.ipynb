```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
print(df.to_string())
g=sb.FacetGrid(df,col='time')
g.map(plt.hist,"tip")
plt.show()
```

```python
df=sb.load_dataset('tips')
#g=sb.FacetGrid(df,col="time",row="sex",hue="smoker")
g=sb.FacetGrid(df,col="time",row="sex",hue="time")
g.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
#statistical estimation
import pandas as pd
import seaborn as sb
from matplotlib import pyplot
df=sb.load_data_set('tips')
sb.barplot(x='sex',y='tip',hue='time',data=df)
pyplot.show()
```

```python
#linear relationship
df=sb.load_dataset('tips')
sb.regplot(x='total_bill',y='tip',data=df)
pyplot.show()
```

```python
#pair plot
sb.pairplot(df)
pyplot.show()
```

```python
df=sb.load_dataset('tips')
selected_vars=['total_bill','tip']
sb.pairplot(df,vars=selected_vars)
pyplot.show()
```

```python
sb.pairplot(df,hue='size')
plt.show()
```

```python
sb.pairplot(df,hue='size',palette='husl')
plt.show()
```

```python
sb.pairplot(df,diag_kind='kde')
plt.show()
```

```python
sb.pairplot(df,kind='reg')
plt.show()
```

```python
#controlling markers
sb.pairplot(df,hue='sex',markers=['o','s'])
plt.show()
```

```python
g=sb.pairplot(df,hue='day')
g.fig.suptitle('Pair Plot of tips datset',y=1.02)
g.set(xticks=[],yticks=[])
plt.show()
```
